# Theano Tutorial: Advanced
We are going to develop **Linear Regression**. It is to predict the target `t` value given `x`, with linear model $y=xw+b$. We use training data to estimate the parameters `w` and `b`, by minimizing error, which is the difference between true target `t` and predicted by model `y`. In equations, we want to minimize error $$E = \frac{1}{n}\sum_{i=1}^n \vert y_i - t_i \vert^2 = \frac{1}{n}\sum_{i=1}^n\vert x_i^T w + b - t_i \vert^2 = \frac{1}{n}\Vert X w + b1 - t\Vert^2$$
where i-th row of matrix $X$ is $x_i^T$, $t$ is the vector of $t_i$'s and $1$ is a column vector whose elements are all 1.

```python
import theano
import theano.tensor as T
import numpy as np
```

Make synthetic training dataset and randomly initialize the parameters

```python
# Create dataset
num_data = 20
xdata = np.linspace(0,10,num_data, dtype=theano.config.floatX)
tdata = np.asarray(0.5*xdata+1+0.5*np.random.randn(num_data), dtype=theano.config.floatX)

# Initialize parameters
wval = float(0.1*np.random.randn(1))
bval = float(0.1*np.random.randn(1))

# Draw data and current model
scatter(xdata,tdata)
plot(xdata, wval*xdata+bval)
```

As expected, the model is not correct for now. We would like to train the model parameters `w` and `b`. In fact, for linear regression problem, analytic solution for the parameters exist. However, for the purpose of exercise, we do Gradient Descent Update in this tutorial. The following calculates gradient of the error function according to the model parameters.

```python
X = T.vector('X') # as we consider 1-d data, it is not matrix but vector
w = theano.shared(wval, name='w')
b = theano.shared(bval, name='b')
y = X*w + b # this one as well. for n-d data, it should be y = T.dot(X,w)+b

t = T.vector('t')
E = T.sum((y-t)**2)/num_data
dEdw = T.grad(E, w)
dEdb = T.grad(E, b)
```

You may have noticed new syntax `shared`. You can think of it as a symbolic variable but has some value. For now, you are ok to say "use `shared` for parameters". However, in fact, it is introduced because of GPU computing. Look up Theano documentation for more details. To review the syntax of `theano.function`, let us draw the same plot by using it.

```python
calc_y = theano.function([X], y)
scatter(xdata,tdata)
plot(xdata, calc_y(xdata))
```

Now coming back to the Gradient descent, let us create function to calculate gradients, and update the parameters at the same time.

```python
lr = 0.003
train = theano.function([], [E,y], updates=[(w, w - lr*dEdw), (b, b - lr*dEdb)], givens=[(X,xdata),(t,tdata)])
```

There are new some new syntax `updates`, `givens`. Whenever the function `train` is called, the `updates` will replace the first shared variable with the second variable for every tuple and `givens` will replace the first expression with the second value for every tuple.

```python
for epoch in xrange(2000):
    error, ydata = train()
    if epoch % 100 == 0: # plot every 100 epoch
        scatter(xdata,tdata)
        plot(xdata, ydata)
        print("error: "+str(error))
        print("w: "+str(w.get_value()))
        print("b: "+str(b.get_value()))
        show()
```

Is it trained well? Try with different `n` and different `lr` or even different dataset.
